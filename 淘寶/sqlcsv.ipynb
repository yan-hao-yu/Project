```python
import pandas as pd
```

```python
dtypes = {'user_id': 'str', 'item_id': 'str', 'behavior_type': 'str', 'user_geohash': 'str', 'item_category': 'str', 'time': 'datetime64'}
```

```python
df = pd.read_excel('data/taobao_data.xlsx', dtype = dtypes)
df.head()
```

```python
print('Record from', df.time.min(), 'to', df.time.max())
```

```python
na_cols = df.isnull().any(axis=0) # 每行有沒有缺失的值
print('NA cols:')
print(na_cols)
print('-'*30)
print('valid records for each cols:')
print(df.isnull().sum()) # 每行中沒有NAN的數量
print('-'*30)
na_rows = df.isnull().any(axis=1) # 每列有沒有缺失的值
print('Total num of NA rows is: {0}'.format(na_rows.sum()))
```

```python
data = df[['user_id', 'item_id', 'behavior_type', 'item_category', 'time']]
data.head(4)
```

```python
data['datetime'] = data['time']
data['date'] = data['time'].dt.date
data['time'] = data['time'].dt.day
data.head(4)
```
